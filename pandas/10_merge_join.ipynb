```python
import pandas as pd
week1 = pd.read_csv("../data/Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("../data/Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("../data/Restaurant - Customers.csv")
foods = pd.read_csv('../data/Restaurant - Foods.csv',index_col='Food ID')
```

```python
week1.head()
```

```python
week2.head()
```

### Inner join and using suffixes parameter

```python
week1.merge(week2,how="inner",
            on="Customer ID",
            suffixes=("_left","_right")
           )
```

### Join on multiple columns

```python
week1.merge(week2, how='inner' ,on=['Customer ID','Food ID'])
```

### outer joins with indicator parameter

```python
week1.merge(week2,how='outer',on='Customer ID').head()
```

```python
week1.merge(week2,
            how='outer',
            on='Customer ID',
            indicator=True).head(10)
```

```python
week1.merge(week2,
            how='outer',
            on='Customer ID',
            indicator=True)['_merge'].value_counts()
```

```python
week1.merge(week2,how="left",on="Customer ID",sort=True).head(4)
```

### using  left_on, right_on parameter

```python
foods.head()
```

```python
customers.head()
```

```python
week1.merge(customers,
            how='left',
            left_on='Customer ID',
            right_on='ID'
           ).drop('Customer ID',axis=1)
```

### Joining based on indexes

```python
week1.merge(foods,how="left",left_on="Food ID",right_index=True)
```

```python
week1.merge(week2,how="left",left_index=True,right_index=True,suffixes=('_week1','_week2'))
```

### pd.merge() method

```python
pd.merge(week1,week2, on = 'Customer ID',how='left').head()
```

```python
pd.merge(week1,foods, left_on = 'Food ID',right_index=True).head()
```
